16. The cleaning dataset is visualize below.

```python
def find_full(findin):
    whattofind = 'Ships From'
    if str(findin).count(whattofind) >0:
        x=str(findin).find(whattofind)
        y= len(whattofind)
        z=str(findin)[x+y:]
        return z
```

```python
def find_state(findin):
    whattofind = ', '
    if str(findin).count(whattofind) >0 :
        x=str(findin).find(whattofind)
        y= len(whattofind)
        z=str(findin)[x+y:].strip()
        #ab3456fd 
#         abfd
        numbers ='0'
        z=''.join(e for e in z if e not in numbers)
        return z
```

```python
def convert(findin):
    whattofind = 'k'
    if str(findin).count(whattofind) >0 :
        z=str(findin)[:len(findin)-1].strip()
#         alphabets ='k'
#         z=''.join(e for e in z if e not in numbers)
        return float(z)*1000
    else:
        return float(findin)
```

```python
df['Full_Origin']=df['description'].apply(find_full)
df['Short_Origin']=np.where(df['china']!= 'Mainland China', df['Full_Origin'].apply(find_state), 'Mainland China')
df['item_sold_int']=df['item sold'].apply(convert)
pd.set_option('display.max_columns', None)
df['origin_num'] = np.where(df['china']!= 'Mainland China', 1, 0)
df['category_num'] = np.where(df['category']!= 'Mobile Phones', 1, 0)
df.head(5)
```
